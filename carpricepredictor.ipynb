```python
def convert_to_int(value):
    try :
      int(value)
      return True
    except ValueError:
        return False
```

```python
def data_cleaning(column):
  for i in range(len(column)):
    if convert_to_int(column[i])==False:
      column[i]=np.nan
    else:
      column[i]=column[i]
    #print(column.type)
  return column


```
