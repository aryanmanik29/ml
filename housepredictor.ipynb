```python
def range(x):
    x=str(x)
    if len(x.split("-"))==2:
        return str((float(x.split("-")[0])+float(x.split("-")[1]))/2)
    try:
        return str(x)
    except:
        return str(x)
```

```python
data2["total_sqft"]=data2["total_sqft"].apply(lambda x:range(x))
```

```python
data2["total_sqft"]=pd.to_numeric(data2["total_sqft"])
```

```python
def convert(x):
  try:
    int(x)
    return True
  except  ValueError:
    return False
```
