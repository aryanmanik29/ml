```python
traindata['tokens'] =traindata['email'].apply(lambda x: x.split())
```

```python
word_columns = pd.DataFrame(traindata['tokens'].tolist(), columns=[f'word_{i+1}' for i in range(traindata['tokens'].str.len().max())])
```

```python
result = pd.concat([traindata, word_columns], axis=1)


result = result.drop(columns=['tokens'])
```
